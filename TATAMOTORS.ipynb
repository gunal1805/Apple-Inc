Full code and Output:

```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv(r"C:\Users\HP\Documents\archive (3)\TATAMOTORS.NS (2).csv")

print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nFirst 5 Rows:")
print(df.head())

print("\nLast 5 Rows:")
print(df.tail())

print("\nInfo:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
df = df.ffill()
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```
