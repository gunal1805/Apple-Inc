TATA:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv(r"C:\Users\HP\Documents\archive (3)\TATAMOTORS.NS (2).csv")

print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

Matplotlib:

```python
# 1 Line Plot – Closing Price Trend
plt.figure()
plt.plot(df["Close"])
plt.title("Tata Motors Closing Price Trend")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()


# 2 Histogram – Close Price Distribution
plt.figure()
plt.hist(df["Close"], bins=20)
plt.title("Distribution of Closing Prices")
plt.xlabel("Close Price")
plt.ylabel("Frequency")
plt.show()


# 3 Scatter Plot – Open vs Close
plt.figure()
plt.scatter(df["Open"], df["Close"])
plt.title("Open vs Close Price")
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.show()


# 4 Bar Plot – First 10 Close Prices
plt.figure()
plt.bar(range(10), df["Close"].head(10))
plt.title("First 10 Closing Prices")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()


# 5 Box Plot – Close Price
plt.figure()
plt.boxplot(df["Close"])
plt.title("Boxplot of Close Price")
plt.ylabel("Close Price")
plt.show()


# 6 Area Plot – Close Price
plt.figure()
plt.fill_between(range(len(df)), df["Close"])
plt.title("Area Plot of Closing Price")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()
```

Seaborn:

```python
# 1 Line Plot
plt.figure()
sns.lineplot(x=df.index, y=df["Close"])
plt.title("Tata Motors Closing Price Trend")
plt.show()


# 2 Histogram
plt.figure()
sns.histplot(df["Close"], bins=20)
plt.title("Distribution of Closing Prices")
plt.show()


# 3 Scatter Plot
plt.figure()
sns.scatterplot(x=df["Open"], y=df["Close"])
plt.title("Open vs Close Price")
plt.show()


# 4 Bar Plot
plt.figure()
sns.barplot(x=df.index[:10], y=df["Close"].head(10))
plt.title("First 10 Closing Prices")
plt.show()


# 5 Box Plot
plt.figure()
sns.boxplot(x=df["Close"])
plt.title("Boxplot of Close Price")
plt.show()


# 6 Area Style Plot
plt.figure()
sns.lineplot(x=df.index, y=df["Close"])
plt.fill_between(df.index, df["Close"], alpha=0.3)
plt.title("Area Style Closing Price Plot")
plt.show()
```
