```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\HP\Documents\archive (1)\weapons_data.csv")

# =============================
# Basic Dataset Analysis
# =============================

print("Shape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nUnique Value Count per Column:")
print(df.nunique())

# =============================
# Select Numeric Columns Only
# =============================

numeric_df = df.select_dtypes(include=['int64','float64'])

print("\nNumeric Columns for Plotting:")
print(numeric_df.columns)

# =============================
# Matplotlib Plots
# =============================

# 1 Line Plot
plt.figure()
plt.plot(numeric_df.iloc[:,0])
plt.title("Line Plot of First Numeric Feature")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()

# 2 Histogram
plt.figure()
plt.hist(numeric_df.iloc[:,0], bins=20)
plt.title("Histogram of First Numeric Feature")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# 3 Scatter Plot
if numeric_df.shape[1] >= 2:
    plt.figure()
    plt.scatter(numeric_df.iloc[:,0], numeric_df.iloc[:,1])
    plt.title("Scatter Plot of First Two Features")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

# 4 Bar Plot
plt.figure()
plt.bar(range(10), numeric_df.iloc[:10,0])
plt.title("Bar Plot of First 10 Values")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()

# 5 Box Plot
plt.figure()
numeric_df.boxplot()
plt.title("Boxplot of Numeric Features")
plt.xticks(rotation=45)
plt.show()

# 6 Area Plot
plt.figure()
plt.fill_between(range(len(numeric_df.iloc[:100,0])), numeric_df.iloc[:100,0])
plt.title("Area Plot of Feature")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()

# =============================
# Seaborn Plots
# =============================

# 1 Line Plot
plt.figure()
sns.lineplot(x=numeric_df.index, y=numeric_df.iloc[:,0])
plt.title("Line Plot of First Numeric Feature")
plt.show()

# 2 Histogram
plt.figure()
sns.histplot(numeric_df.iloc[:,0], bins=20)
plt.title("Histogram of First Numeric Feature")
plt.show()

if numeric_df.shape[1] >= 2:
    plt.figure()
    sns.scatterplot(x=numeric_df.iloc[:,0], y=numeric_df.iloc[:,1])
    plt.title("Scatter Plot of First Two Features")
    plt.show()

plt.figure()
sns.barplot(x=numeric_df.index[:10], y=numeric_df.iloc[:10,0])
plt.title("Bar Plot of First 10 Values")
plt.show()

plt.figure()
sns.boxplot(data=numeric_df)
plt.title("Boxplot of Numeric Features")
plt.xticks(rotation=45)
plt.show()

plt.figure()
sns.lineplot(x=numeric_df.index[:100], y=numeric_df.iloc[:100,0])
plt.fill_between(numeric_df.index[:100], numeric_df.iloc[:100,0], alpha=0.3)
plt.title("Area Style Plot")
plt.show()
```
