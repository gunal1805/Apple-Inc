```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\HP\Documents\gold dataset\FINAL_USO.csv")

print("Shape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nUnique Value Count per Column:")
print(df.nunique())
```
