Problem Statement:
This dataset presents a supervised multi-class classification problem focused on detecting and classifying cyber-attacks in IoT network environments using over one million network traffic records with 47 statistical, protocol-level, and flow-based features. Each record contains attributes such as flow duration, packet rates, protocol types (TCP, UDP, ICMP), TCP flag counts, packet size metrics, and statistical measures like mean and variance, with a corresponding label identifying normal traffic or specific attack types such as DDoS and DoS variants. The objective is to develop a machine learning model that can accurately distinguish between normal and malicious traffic, effectively classify different attack categories, and minimize false negatives while maintaining high precision, thereby enabling reliable and automated intrusion detection in IoT-based systems.
