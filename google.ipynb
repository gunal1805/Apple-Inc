Problem statement:

The objective of this project is to analyze the Google Play Store applications dataset (apps.csv) along with the corresponding user feedback dataset (user_reviews.csv) to understand the factors influencing app popularity, ratings, and user satisfaction. The study aims to explore relationships between app features such as category, size, price, installs, and content rating, and compare them with user sentiments, reviews, and feedback patterns. By performing exploratory data analysis and applying suitable data analytics or machine learning techniques, the project seeks to identify trends that impact app success, detect key drivers of positive and negative reviews, and generate insights that can help developers improve app performance and user experience in the competitive mobile application market

apps.csv:

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\Documents\archive\apps.csv")

print("Shape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())
df = df.ffill()
print(df.isnull().sum())


print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nUnique Value Count per Column:")
print(df.nunique())
```
