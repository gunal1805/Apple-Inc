Plots for Google dataset:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
apps_df = pd.read_csv(r"C:\Users\HP\Documents\archive\apps.csv")

# Check if dataset loaded
print(apps_df.head())
```

Histogram:

```python
plt.figure()
plt.hist(apps_df["Rating"].dropna(), bins=20)
plt.title("App Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Apps")
plt.show()
```

Bar Plot:

```python
plt.figure()
apps_df["Category"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 App Categories")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.show()
```

Scatter Plot:

```python
plt.figure()
plt.scatter(apps_df["Reviews"], apps_df["Rating"])
plt.title("Rating vs Reviews")
plt.xlabel("Reviews")
plt.ylabel("Rating")
plt.show()
```

Box Plot:

```python
plt.figure()
apps_df.boxplot(column="Rating", by="Category", rot=90)
plt.title("Rating by Category")
plt.show()
```

```python
Pie chart:
```

```python
plt.figure()
apps_df["Type"].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.title("Free vs Paid Apps")
plt.ylabel("")
plt.show()
```

Bar Plot:

```python
reviews_df = pd.read_csv(r"C:\Users\HP\Documents\archive\user_reviews.csv")

plt.figure()
reviews_df["Sentiment"].value_counts().plot(kind="bar")
plt.title("User Review Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

Seaborn:

Histogram:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()

sns.histplot(apps_df["Rating"], bins=20)

plt.title("App Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()
```

Scatter plot:

```python
plt.figure()

sns.scatterplot(x="Reviews", y="Rating", data=apps_df)

plt.title("Reviews vs Rating")

plt.show()
```

Bar plot

```python
plt.figure()

top_categories = apps_df["Category"].value_counts().head(10)

sns.barplot(x=top_categories.index, y=top_categories.values)

plt.title("Top 10 App Categories")
plt.xticks(rotation=45)

plt.show()
```

Box plot:

```python
plt.figure()

sns.boxplot(y=apps_df["Rating"])

plt.title("Rating Spread")

plt.show()
```

Pie plot:

```python
plt.figure()

apps_df["Content Rating"].value_counts().head().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Content Rating Distribution")

plt.ylabel("")

plt.show()
```

Sentiment Distribution – User Reviews:

```python
plt.figure()

sns.countplot(x="Sentiment", data=reviews_df)

plt.title("User Review Sentiment Distribution")

plt.show()
```
