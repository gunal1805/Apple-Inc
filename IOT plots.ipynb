IOT plot:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\HP\Documents\archive (5)\IoT_Intrusion.csv")

print("Shape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns)
```

Matplotlib:

```python
# 1 Histogram – Feature Distribution
plt.figure()
plt.hist(df.iloc[:,0], bins=20)
plt.title("Feature Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()


# 2 Line Plot
plt.figure()
plt.plot(df.iloc[:100,0])
plt.title("Feature Trend")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()


# 3 Scatter Plot
plt.figure()
plt.scatter(df.iloc[:100,0], df.iloc[:100,1])
plt.title("Feature 1 vs Feature 2")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()


# 4 Bar Plot
plt.figure()
df.iloc[:10,0].plot(kind="bar")
plt.title("First 10 Values")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()


# 5 Box Plot
plt.figure()
plt.boxplot(df.iloc[:,0])
plt.title("Boxplot of Feature")
plt.show()


# 6 Area Plot
plt.figure()
plt.fill_between(range(len(df.iloc[:100,0])), df.iloc[:100,0])
plt.title("Area Plot of Feature")
plt.show()
```

Seaborn:

```python
# 1 Histogram
plt.figure()
sns.histplot(df.iloc[:,0], bins=20)
plt.title("Feature Distribution")
plt.show()


# 2 Line Plot
plt.figure()
sns.lineplot(x=df.index[:100], y=df.iloc[:100,0])
plt.title("Feature Trend")
plt.show()


# 3 Scatter Plot
plt.figure()
sns.scatterplot(x=df.iloc[:100,0], y=df.iloc[:100,1])
plt.title("Feature 1 vs Feature 2")
plt.show()


# 4 Bar Plot
plt.figure()
sns.barplot(x=df.index[:10], y=df.iloc[:10,0])
plt.title("First 10 Values")
plt.show()


# 5 Box Plot
plt.figure()
sns.boxplot(x=df.iloc[:,0])
plt.title("Boxplot of Feature")
plt.show()


# 6 Area Style Plot
plt.figure()
sns.lineplot(x=df.index[:100], y=df.iloc[:100,0])
plt.fill_between(df.index[:100], df.iloc[:100,0], alpha=0.3)
plt.title("Area Style Plot")
plt.show()
```
