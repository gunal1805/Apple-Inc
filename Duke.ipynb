Full code and Output:

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\Documents\duke dataset\duke-breast-cancer.txt")

print("Shape of dataset:")
print(df.shape)

print("\nColumn names:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nUnique Value Count per Column:")
print(df.nunique())
```

Summary:

The Duke Breast Cancer dataset contains gene expression data collected from breast cancer patients. Each row represents a patient sample, while the columns represent gene expression levels. This dataset helps researchers analyze how genes behave in cancerous tissues and identify patterns that may help in diagnosing or predicting breast cancer. Through exploratory data analysis, we can examine the dataset structure, detect missing values, and understand statistical properties of gene expression data, which can support further research in medical data analysis and cancer prediction.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1 Line Plot – Closing Price Trend
plt.figure()
plt.plot(df["Close"])
plt.title("Closing Price Trend")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()


# 2 Histogram – Close Price Distribution
plt.figure()
plt.hist(df["Close"], bins=20)
plt.title("Close Price Distribution")
plt.xlabel("Close Price")
plt.ylabel("Frequency")
plt.show()


# 3 Scatter Plot – Open vs Close
plt.figure()
plt.scatter(df["Open"], df["Close"])
plt.title("Open vs Close Price")
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.show()


# 4 Bar Plot – First 10 Close Prices
plt.figure()
plt.bar(range(10), df["Close"].head(10))
plt.title("First 10 Closing Prices")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()


# 5 Box Plot – Close Price
plt.figure()
plt.boxplot(df["Close"])
plt.title("Boxplot of Close Price")
plt.ylabel("Close Price")
plt.show()


# 6 Area Plot – Close Price
plt.figure()
plt.fill_between(range(len(df)), df["Close"])
plt.title("Area Plot of Close Price")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()
```
