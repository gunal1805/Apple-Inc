Matplotlib:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
plt.plot(df["Close"])
plt.title("Closing Price Trend")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()



plt.figure()
plt.hist(df["Close"], bins=20)
plt.title("Close Price Distribution")
plt.xlabel("Close Price")
plt.ylabel("Frequency")
plt.show()


plt.figure()
plt.scatter(df["Open"], df["Close"])
plt.title("Open vs Close Price")
plt.xlabel("Open Price")
plt.ylabel("Close Price")
plt.show()


plt.figure()
plt.bar(range(10), df["Close"].head(10))
plt.title("First 10 Closing Prices")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()


plt.figure()
plt.boxplot(df["Close"])
plt.title("Boxplot of Close Price")
plt.ylabel("Close Price")
plt.show()


plt.figure()
plt.fill_between(range(len(df)), df["Close"])
plt.title("Area Plot of Close Price")
plt.xlabel("Index")
plt.ylabel("Close Price")
plt.show()
```
