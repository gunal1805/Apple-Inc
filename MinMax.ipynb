```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv(r"C:\Users\HP\Documents\archive (3)\TATAMOTORS.NS (2).csv")
scaler = MinMaxScaler()
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = scaler.fit_transform(df[[col]])
df
```
